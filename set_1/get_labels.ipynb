```python
import matplotlib.pyplot as plt
import numpy as np

data = []
with open('extracted_labels.txt', 'r') as f:
    for line in f:
        data.append(int(line.strip()))

x = np.arange(len(data)) # generate a sequence of x values
colors = ['red', 'green', 'blue', 'orange'] # define a list of colors for each class
y = np.array([colors[value] for value in data]) # generate a list of colors for each point based on its class

fig, ax = plt.subplots(figsize=(40, 5)) # set the size of the plot
ax.scatter(x, y, s=10) # create the scatter plot
ax.set_title('Sequence of Values')
ax.set_xlabel('Line Number')
ax.set_ylabel('Value')
ax.set_yticks(colors) # set the y-axis ticks to be the class colors
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = []
with open('extracted_labels.txt', 'r') as f:
    for line in f:
        data.append(int(line.strip()))

x = np.arange(len(data)) # generate a sequence of x values
colors = ['red', 'green', 'blue', 'orange'] # define a list of colors for each class
y = np.array([value for value in data]) # generate a list of class values for each point

fig, ax = plt.subplots(figsize=(40, 5)) # set the size of the plot
for i in range(len(colors)): # loop over each class color
    mask = y == i # create a mask for values that belong to this class
    ax.scatter(x[mask], y[mask], c=colors[i], s=10) # create a scatter plot for this class

ax.set_title('Sequence of Values')
ax.set_xlabel('Line Number')
ax.set_ylabel('Value')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = []
with open('extracted_labels.txt', 'r') as f:
    for line in f:
        data.append(int(line.strip()))

x = np.arange(len(data)) # generate a sequence of x values
colors = ['red', 'green', 'blue', 'orange'] # define a list of colors for each class
y = np.array([value for value in data]) # generate a list of class values for each point

fig, ax = plt.subplots(figsize=(40, 5)) # set the size of the plot
for i in range(len(colors)): # loop over each class color
    mask = y == i # create a mask for values that belong to this class
    ax.scatter(x[mask], y[mask], c=colors[i], s=10) # create a scatter plot for this class

ax.set_title('SEQUENCE OF PACKETS AND THEIR LABELS')
ax.set_xlabel('Packets')
ax.set_yticks(range(4)) # set the y-ticks to the class values
ax.set_yticklabels(['Benign', 'Attack', 'Infection', 'Reconnais.']) # set the y-tick labels to the class names
plt.show()
```
